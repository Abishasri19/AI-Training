```python
import soundfile as sf
import torch
import numpy as np
import librosa
from datasets import load_dataset
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor, Wav2Vec2Model
# load pretrained model
#The facebook model will not work because there is no processor
#processor = Wav2Vec2Processor.from_pretrained("facebook/wav2vec2-large-xlsr-53")
processor = Wav2Vec2Processor.from_pretrained("jonatasgrosman/wav2vec2-large-xlsr-53-english")
model = Wav2Vec2ForCTC.from_pretrained("jonatasgrosman/wav2vec2-large-xlsr-53-english")
#model = Wav2Vec2ForCTC.from_pretrained("facebook/wav2vec2-large-xlsr-53")
```

```python
datasets=load_dataset('librispeech_asr','clean')
```

```python
test_aud=datasets["test"]["audio"]
test_fil=datasets["test"]["text"]
```

```python
audio_sig=[]
for i in test_aud:
  audio_sig.append(i['array'])
audio_sig=np.array(audio_sig)
audio_sig
```

```python
predicted_sentences=[]
for i in range(0,10):
  inputs = processor(audio_sig[i], sampling_rate=16000, return_tensors="pt")
  with torch.no_grad():
    logits = model(inputs.input_values).logits
  predicted_ids = torch.argmax(logits, dim=-1)
  predicted_sentence = processor.batch_decode(predicted_ids)
  print(predicted_sentence)
  predicted_sentences.append(predicted_sentence)
  #predicted_ids = torch.argmax(logits, dim=-1)
  #predicted_sentences = processor.batch_decode(predicted_ids)
  #print(predicted_sentences)
```

```python
full_pred_sentences=[]
for i in range(0,2620):
  inputs = processor(audio_sig[i], sampling_rate=16000, return_tensors="pt")
  with torch.no_grad():
    logits = model(inputs.input_values).logits
  predicted_ids = torch.argmax(logits, dim=-1)
  predicted_sentence = processor.batch_decode(predicted_ids)
  print(predicted_sentence)
  full_pred_sentences.append(predicted_sentence)
```
