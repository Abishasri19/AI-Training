```python
!mkdir test_Flac
```

```python
!tar -C test_Flac -xvzf test-clean.tar.gz
```

```python
dirs=os.listdir("test_Flac/LibriSpeech/test-clean")
dir=list(sorted(set(dirs)))
print(dir)
```

```python
!mkdir Test_WAV
```

```python
#Load .flac files and convert to .wav
for i in dir:
  ids=os.listdir("test_Flac/LibriSpeech/test-clean/"+i)
  for j in ids:
    files=os.listdir("test_Flac/LibriSpeech/test-clean/"+i+"/"+j)
    for fil in files:
      if(fil[-4:]=="flac"):
        os.system('ffmpeg -i test_Flac/LibriSpeech/test-clean/'+i+'/'+j+'/'+fil+' Test_WAV/'+fil[:-5]+'.wav')
```
